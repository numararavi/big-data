```python
from collections import defaultdict

def mapper(line):
    """
    Fungsi mapper untuk memecah data dan menghitung jumlah nilai dan banyaknya data.

    Args:
        line (str): Satu baris data dari file CSV.

    Yields:
        tuple: Tuple berisi (1, (nilai, 1))
    """
    _, nilai = line.strip().split(',')
    yield (1, (int(nilai), 1))

def reducer(key, values):
    """
    Fungsi reducer untuk menghitung rata-rata.

    Args:
        key (int): Kunci (dalam kasus ini selalu 1).
        values: Iterable dari nilai (nilai, 1).

    Yields:
        tuple: Tuple berisi (key, rata-rata)
    """
    total_nilai = 0
    jumlah_data = 0
    for nilai, count in values:
        total_nilai += nilai
        jumlah_data += count
    yield (key, total_nilai / jumlah_data)

def hitung_rata_rata_mapreduce(nama_file):
    """
    Fungsi utama untuk menghitung rata-rata menggunakan MapReduce.

    Args:
        nama_file (str): Nama file CSV.

    Returns:
        float: Rata-rata nilai.
    """

    intermediate = defaultdict(list)
    with open(nama_file, 'r') as file:
        for line in file:
            for key, value in mapper(line):
                intermediate[key].append(value)

    result = []
    for key in intermediate:
        for value in reducer(key, intermediate[key]):
            result.append(value)

    return result[0][1]

# Contoh penggunaan
rata_rata = hitung_rata_rata_mapreduce('nilai_ujian.csv')
print(f"Rata-rata nilai: {rata_rata:.2f}")
```

```python
import csv
from collections import defaultdict

# Fungsi untuk membaca file CSV dan memproses datanya
def baca_file_csv(nama_file):
    with open(nama_file, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Lewati header
        data = list(csv_reader)
    return data

# Fungsi Map
def map_values(input_data):
    mapped = []
    for row in input_data:
        jam_belajar, nilai = row
        mapped.append((int(jam_belajar.strip()), int(nilai.strip())))
    return mapped

# Fungsi Reduce
def reduce_values(mapped_data):
    nilai_total = defaultdict(int)
    jumlah_siswa = defaultdict(int)
    
    for jam_belajar, nilai in mapped_data:
        nilai_total[jam_belajar] += nilai
        jumlah_siswa[jam_belajar] += 1
    
    # Hitung rata-rata
    rata_rata_nilai = {}
    for jam_belajar in nilai_total:
        rata_rata_nilai[jam_belajar] = nilai_total[jam_belajar] / jumlah_siswa[jam_belajar]
    
    return rata_rata_nilai

# Proses utama
def hitung_rata_rata(nama_file):
    # Membaca data dari file CSV
    input_data = baca_file_csv(nama_file)
    
    # Proses MapReduce
    mapped_data = map_values(input_data)
    rata_rata = reduce_values(mapped_data)
    
    # Cetak hasil rata-rata nilai per jam belajar
    for jam_belajar, rata in rata_rata.items():
        print(f"Rata-rata nilai untuk {jam_belajar} jam belajar: {rata:.2f}")

# Jalankan fungsi utama dengan nama file CSV
hitung_rata_rata('nilai_ujian.csv')
```
